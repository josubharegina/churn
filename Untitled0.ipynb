```python
dataset['avg_transaction_value'] = pd.to_numeric(dataset['avg_transaction_value'], errors='coerce')
dataset['avg_transaction_value'].fillna(dataset['avg_transaction_value'].median(), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

# These are the actual categorical columns present in your dataset
categorical_cols = [
    'gender', 'region_category', 'membership_category',
    'joined_through_referral', 'preferred_offer_types',
    'medium_of_operation', 'internet_option', 'used_special_discount',
    'offer_application_preference', 'past_complaint',
    'complaint_status', 'feedback'
]

for col in categorical_cols:
    # Fill missing values with a placeholder string first so the encoder doesn't crash
    dataset[col] = dataset[col].fillna('Unknown')
    dataset[col] = labelencoder.fit_transform(dataset[col])
```

```python
from sklearn.model_selection import train_test_split

# Drop identifier columns and the target column 'feedback' from features
X = dataset.drop(['customer_id', 'Name', 'feedback'], axis=1)

# Set 'feedback' as your target variable
y = dataset['feedback']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import train_test_split

# Clean and update the numeric columns
dataset['points_in_wallet'] = pd.to_numeric(dataset['points_in_wallet'], errors='coerce')
dataset['points_in_wallet'] = dataset['points_in_wallet'].fillna(dataset['points_in_wallet'].median())

dataset['avg_frequency_login_days'] = pd.to_numeric(dataset['avg_frequency_login_days'], errors='coerce')
dataset['avg_frequency_login_days'] = dataset['avg_frequency_login_days'].fillna(dataset['avg_frequency_login_days'].median())

# Separate features and target
X = dataset.drop(['customer_id', 'Name', 'security_no', 'joining_date', 'referral_id', 'last_visit_time', 'feedback'], axis=1)
y = dataset['feedback']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```
